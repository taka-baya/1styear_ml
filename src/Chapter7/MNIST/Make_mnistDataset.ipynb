```python
# Pytorchに関するライブラリをインポート
import torch
from torchvision import datasets, transforms

# 配列操作を行うためにnumpyをインポート
import numpy as np

# 画像表示などを行うためにmatplotlibをインポート
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Totenser関数でtensor化
# Normalise関数で正規化(-1.0〜1.0をとるような値として変換)
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, ), (0.5, ))])
```

```python
# 変数datasetにMNISTのデータセットを格納
# ダウンロードしたデータセットは、./dataフォルダに配置される。
dataset = datasets.MNIST('./data', train=False, download=True, transform=transform, target_transform=None)
```

```python
# 変数datasetの画像データの型を確認する
print(type(dataset[0][0]))
```

```python
# 変数datasetの画像データの型を確認する
print(type(dataset[0][1]))
```
